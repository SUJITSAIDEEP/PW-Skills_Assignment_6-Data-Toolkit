#THEORY

#1.What is NumPy, and why is it widely used in Python ?

-NumPy is a Python library used for fast numerical computations and handling large multidimensional arrays efficiently.
It is widely used because it provides high-performance mathematical operations and forms the foundation for many data science and machine learning libraries.

#2.How does broadcasting work in NumPy ?

-Broadcasting in NumPy is a feature that allows operations between arrays of different shapes by automatically expanding the smaller array to match the larger one’s shape.
It simplifies code and avoids creating unnecessary copies of data while performing element-wise operations.

#3.What is a Pandas DataFrame ?
-A Pandas DataFrame is a two-dimensional, tabular data structure in Python with labeled rows and columns, similar to an Excel sheet.
It allows easy data manipulation, analysis, and cleaning using powerful built-in functions.

#4.Explain the use of the groupby() method in Pandas ?
-The groupby() method is used to split data into groups based on one or more columns and apply functions like sum, mean, or count.
It helps in performing efficient aggregation and analysis on grouped data.

#5.Why is Seaborn preferred for statistical visualizations ?
-Seaborn is preferred because it provides high-level, easy-to-use functions for creating attractive and informative statistical plots.
It integrates well with Pandas and automatically handles themes, color palettes, and statistical calculations.

#6.What are the differences between NumPy arrays and Python lists ?
-NumPy arrays are homogeneous, fixed-size, and support fast mathematical operations.
Python lists are heterogeneous, flexible, but slower for numerical computations.

#7.What is a heatmap, and when should it be used ?
-A heatmap displays data values as colors in a matrix form.
It is used to visualize correlations or intensity patterns between variables.

#8.What does the term “vectorized operation” mean in NumPy ?
-Vectorized operations allow performing arithmetic or logical operations on entire arrays without using loops.
This makes computation faster and more concise.

#9.How does Matplotlib differ from Plotly ?
-Matplotlib creates static, publication-quality plots.
Plotly creates interactive, web-based visualizations.

#10.What is the significance of hierarchical indexing in Pandas ?
-Hierarchical indexing (MultiIndex) allows multiple index levels on rows or columns.
It simplifies handling high-dimensional data and complex grouping operations.

#11.What is the role of Seaborn’s pairplot() function ?
-The pairplot() function plots pairwise relationships between numerical columns in a dataset.
It helps visualize correlations and patterns between multiple variables.

#12.What is the purpose of the describe() function in Pandas ?
-The describe() function generates summary statistics of numerical data.
It provides measures like mean, std, min, and quartiles at once.

#13.Why is handling missing data important in Pandas ?
-Handling missing data ensures accuracy and reliability of analysis.
Ignoring it can lead to biased or incorrect results.

#14.What are the benefits of using Plotly for data visualization ?
-Plotly allows creating highly interactive and visually appealing charts.
It supports 3D plotting, real-time updates, and web integration.

#15.How does NumPy handle multidimensional arrays ?
-NumPy represents multidimensional data using the ndarray object.
It stores data in a contiguous memory block, allowing efficient indexing and computation.

#16.What is the role of Bokeh in data visualization ?
-Bokeh is used for creating interactive and web-ready visualizations in Python.
It allows zooming, panning, and real-time updates for large datasets.

#17.Explain the difference between apply() and map() in Pandas ?
-map() is used for element-wise transformations on a Series.
apply() works on both Series and DataFrames to apply custom functions along rows or columns.

#18.What are some advanced features of NumPy ?
-Advanced NumPy features include broadcasting, linear algebra, random number generation, and FFT.
It also supports memory mapping and integration with C/C++ & Python Code.

#19.How does Pandas simplify time series analysis ?
-Pandas provides built-in support for date-time indexing, resampling, and frequency conversion.
It makes time-based data handling and trend analysis very efficient.

#20.What is the role of a pivot table in Pandas ?
-A pivot table summarizes and reorganizes data based on columns and aggregation functions.
It helps in quick multi-dimensional data analysis.

#21.Why is NumPy’s array slicing faster than Python’s list slicing ?
-NumPy arrays store data in contiguous memory blocks.
This allows faster access and slicing compared to Python lists that store references.

#22.What are some common use cases for Seaborn ?
-Seaborn is used for visualizing distributions, correlations, and categorical data.
Common plots include heatmaps, boxplots, pairplots, and bar plots.

```python
#1. How do you create a 2D NumPy array and calculate the sum of each row ?
import numpy as np
arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sum = arr.sum(axis=1)
print(row_sum)  # Output: [6 15]
```

```python
#2. Write a Pandas script to find the mean of a specific column in a DataFrame.
import pandas as pd
data = {'Marks': [80, 90, 75, 85]}
df = pd.DataFrame(data)
mean_value = df['Marks'].mean()
print(mean_value)
```

```python
#3. Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 7, 8]

plt.scatter(x, y)
plt.title("Simple Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap ?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame with varied values
df = pd.DataFrame({
    'A': [1, 5, 3, 7, 2],
    'B': [2, 6, 4, 8, 1],
    'C': [10, 20, 15, 30, 25]
})

# Calculate correlation
corr = df.corr()

# Heatmap with color scale
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Heatmap")
plt.show()
```

```python
#5. Generate a bar plot using Plotly.
import plotly.express as px

# Sample data
data = {
    'Fruits': ['Apple', 'Banana', 'Orange', 'Mango'],
    'Quantity': [10, 15, 7, 12]
}

# Create bar plot
fig = px.bar(data, x='Fruits', y='Quantity', title='Fruit Quantity')
fig.show()
```

```python
#6. Create a DataFrame and add a new column based on an existing column.
import pandas as pd

# Create DataFrame
data = {'Marks': [80, 90, 75, 85]}
df = pd.DataFrame(data)

# Add new column 'Grade' based on 'Marks'
df['Grade'] = df['Marks'].apply(lambda x: 'A' if x >= 85 else 'B')

print(df)
```

```python
#7.Write a program to perform element-wise multiplication of two NumPy arrays.
import numpy as np

# Create two arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Element-wise multiplication
result = a * b
print(result)
```

```python
#8.Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

# Plot multiple lines
plt.plot(x, y1, label='Line 1', color='blue', marker='o')
plt.plot(x, y2, label='Line 2', color='red', marker='s')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plot')
plt.legend()  # Show legend
plt.grid(True)
plt.show()
```

```python
#9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
import pandas as pd

# Create DataFrame
data = {'Name': ['Amit', 'Sita', 'Ravi', 'Maya'],
        'Marks': [85, 92, 78, 88]}
df = pd.DataFrame(data)

# Filter rows where Marks > 80
filtered_df = df[df['Marks'] > 80]
print(filtered_df)
```

```python
#10.Create a histogram using Seaborn to visualize a distribution.
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = [12, 15, 12, 18, 19, 15, 17, 20, 22, 18, 15, 19]

# Create histogram
sns.histplot(data, bins=5, kde=True, color='green')
plt.title("Histogram of Sample Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#11.Perform matrix multiplication using NumPy.
import numpy as np

# Create two matrices
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

# Matrix multiplication
result = A @ B   # or use np.dot(A, B)
print(result)
```

```python
#13. Create a 3D scatter plot using Plotly.
import plotly.express as px
import pandas as pd

# Spread out sample data
data = {
    'X': [10, 50, 30, 80, 60],
    'Y': [100, 150, 120, 180, 160],
    'Z': [50, 200, 70, 250, 90],
    'Category': ['A', 'B', 'A', 'B', 'A']
}

df = pd.DataFrame(data)

# 3D scatter plot
fig = px.scatter_3d(df, x='X', y='Y', z='Z', color='Category', size='Z', title='3D Scatter Plot')
fig.show()

```
